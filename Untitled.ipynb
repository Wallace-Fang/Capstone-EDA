```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import dask.dataframe as dd
```

```python
taobao_df = pd.read_csv('taobao.csv')
```

```python
taobao_df.head(10)
```

I have a feeling certain occupations would take advantage of a good discount. The Taobao dataset has a column for occupations that lists the user as either a student or not, 1 for student and 0 for not. College students would probably make up only a percentage of the total userbase, so I would need to refine the dataset to only include college students to see the distribution of their shopping levels. 

```python
college_student_only = taobao_df[taobao_df['occupation']== 1]
non_college_student_only = taobao_df[taobao_df['occupation']== 0]
```

```python
#sns.histplot(data=college_student_only, x='shopping_level')
plt.pie(taobao_df['occupation'].value_counts(), labels = ['non_college student', 'college student'], autopct='%1.1f%%')
```

As expected, college students only made up a small percentage of taobao's userbase with 5.45%, but i still wonder about their shopping habits in comparsion to rest of the userbase.

```python
sns.histplot(data=taobao_df, x='occupation', hue='shopping_level', bins=2, hue_order=['1', '2', '3'])
```

The non-college students seems to have a more varied shopping level, even though the majority of them are still level 3 shoppers (the highest). In comparsion, at first glance, the college students all seemed to be level 3 shoppers but upon closer inspection you see some bold black lines underneth the sole colored bar. Lets have a closer inspection of the distribution of the shopping levels of the college students. 

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,10))
labels = ['level 3', 'level 2', 'level 1']

ax1.pie(non_college_student_only['shopping_level'].value_counts(), labels = labels, autopct='%1.1f%%')
ax1.set_title('Non College Student Shopping Levels')

ax2.pie(college_student_only['shopping_level'].value_counts(), labels = labels, autopct='%1.1f%%')
ax2.set_title('College Students Shopping Levels')
```

As we can see college students did have other shopping levels other than 3, it was just difficult to see in the bar graph since the level 3 bar took up most of the space. I expected college students to do a bunch of shopping, especially when there is a good deal (since most are limited in their budget), but I did not expected most of them to be heavy shoppers. From the two pie graphs we see, 93% of students were heavy shoppers, whereas only 84.7% of the non-college students are heavy shoppers. But does this correspond to the ad clicks?

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,10))
labels = ['not_clicked', 'clicked']

ax1.pie(non_college_student_only['clk'].value_counts(), labels=labels, autopct='%1.1f%%')
ax1.set_title('% of Non College Student Clicking on Ads')

ax2.pie(college_student_only['clk'].value_counts(), labels =labels, autopct='%1.1f%%')

ax2.set_title('% of College Student Clicking on Ads')
```

The % of individuals clicking on ads are quite close, not what I was expecting given college students had a higher shopping level than the general public. Perhaps some other facts needs to be considered.
